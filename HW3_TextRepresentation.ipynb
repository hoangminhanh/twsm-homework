```python
import pickle
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import Binarizer
from gensim.corpora import Dictionary
from gensim.models import TfidfModel
import numpy as np
```

```python
# Importing the Stemmed data from Google Drive

data_path = "/content/drive/MyDrive/TWSM"
infile = open(data_path+'/Stemmed.pkl','rb')
data = pickle.load(infile)

data_stemmed = pd.DataFrame(data, columns=["stemmed_text"])
data_stemmed.head()
```

BAG-OF-WORDS

```python
# Counting the the absolute number of apearraces of a word
absolute = CountVectorizer(max_df=0.95, min_df=0.05)

# Calculating the relative frequency of a word 
relative = TfidfVectorizer(max_df=0.95, min_df=0.05,use_idf=False, norm='l1')

# Calculating 
tf_idf = TfidfVectorizer(max_df=0.95, min_df=0.05, smooth_idf=False)
```

```python
absolute_stemmed = absolute.fit_transform(data)
relative_stemmed = relative.fit_transform(data)
tf_idf_stemmed = tf_idf.fit_transform(data)
```

```python
absolute_df = pd.DataFrame(absolute_stemmed.toarray(), columns=absolute.get_feature_names_out())
absolute_df.head()
```

```python
relative_df = pd.DataFrame(relative_stemmed.toarray(), columns=relative.get_feature_names_out())
relative_df.head()
```

```python
tf_idf_df = pd.DataFrame(tf_idf_stemmed.toarray(), columns=tf_idf.get_feature_names_out())
tf_idf_df.head()
```

```python
binarizer = Binarizer()

absolute_stemmed_bi = binarizer.fit_transform(absolute_stemmed)

print(absolute_stemmed.toarray().max())
print(absolute_stemmed_bi.toarray().max())
```

```python
first_doc_df = pd.DataFrame(np.array(np.unique(data_stemmed.iloc[0,0].split())).T, columns=["keys"]) # Splitting the stemmed text of the first document into separate words
first_doc_df = pd.merge(first_doc_df, absolute_df.iloc[0,:], how="left", left_on="keys", right_index=True)
first_doc_df = pd.merge(first_doc_df, relative_df.iloc[0,:], how="left", left_on="keys", right_index=True)
first_doc_df = pd.merge(first_doc_df, tf_idf_df.iloc[0,:], how="left", left_on="keys", right_index=True)
first_doc_df = pd.merge(first_doc_df, pd.DataFrame(absolute_stemmed_bi.toarray(), columns=absolute.get_feature_names()).iloc[0,:], how="left", left_on="keys", right_index=True)

first_doc_df.columns = ["keys", "absolute", "relative", "TF-IDF", "OHE"]
first_doc_df.fillna(0, inplace=True)
first_doc_df.head()
```

```python
# Sort the df based on descending tf-idf

first_doc_df.sort_values("TF-IDF", ascending=False).head()
```

```python
# Sort the df based on descending absolute frequency

first_doc_df.sort_values("absolute", ascending=False).head()
```

BAG-OF-WORDS (Gensim)

```python
corpus_gen = [doc.split() for doc in data]
```

```python
from gensim import corpora
id2word = corpora.Dictionary(corpus_gen)
```

```python
id2word.filter_extremes(no_below=566, no_above=0.95)

# This helps filter out tokens based on their frequency. In this case, it is tokens that are contained in at least 566 documents, and no more than 95% document
```

```python
print(id2word.token2id) # token id

print(id2word.token2id.keys())

print(id2word.dfs) # document frequency = how many documents contain this token
```

```python
# Convert document into the bag-of-words (BoW) format = list of (token_id, token_count) tuples.

corpus1=[id2word.doc2bow(doc) for doc in corpus_gen]

corpus1[0]
```

```python
# list of (token, relative frequency of token) tuples  

corpus2=[[(token[0],(token[1]/sum(n for _, n in doc))) for token in doc] for doc in corpus1]

corpus2[0]
```

```python
# List of tokens that only appear once 
corpus3=[[(token[0],1) for token in doc] for doc in corpus1]

corpus3[0]
```

```python
tfidf=TfidfModel(dictionary=id2word, normalize=True)
corpus4=[tfidf[id2word.doc2bow(doc)] for doc in corpus_gen]
```
