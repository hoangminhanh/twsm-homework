```python
import pickle
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import Binarizer
import numpy as np
```

```python
# Importing the stemmed News Group data

from google.colab import drive 
drive.mount("/content/drive")

data_path = "/content/drive/MyDrive/TWSM"
infile = open(data_path+'/Stemmed.pkl','rb')
data = pickle.load(infile)

stem = pd.DataFrame(data, columns=["stemmed_text"])
stem.head()
```

```python
# Importing the lemmatized News Group data

infile = open(data_path+'/Lemma.pkl','rb')
data = pickle.load(infile)

lemma = pd.DataFrame(data, columns=["lemmatized_text"])
lemma.head()
```

```python
# Importing the dataset to get the targets

df = pd.read_json('https://raw.githubusercontent.com/selva86/datasets/master/newsgroups.json')
df.head()
```

```python
# Examine the dataset

print('Possible tagret values:')
print(df.target_names.unique())
print(' ')
print('Class distribution:')
print(df.target_names.value_counts())
```

```python
# Combining the stemmed and lemmatized text with their targets 

stem_df = pd.concat([stem, df], axis=1).drop(["content"], axis=1)

lemma_df = pd.concat([lemma, df], axis=1).drop(["content"], axis=1)
```

```python
stem_df.head()
```

```python
lemma_df.head()
```

```python
# Train test split 

from sklearn.model_selection import train_test_split

X_stem = stem_df.stemmed_text
y_stem = stem_df.target

X_train_stem, X_test_stem, y_train_stem, y_test_stem = train_test_split(X_stem, y_stem, test_size=0.1)

X_lemma = lemma_df.lemmatized_text
y_lemma = lemma_df.target

X_train_lemma, X_test_lemma, y_train_lemma, y_test_lemma = train_test_split(X_lemma, y_lemma, test_size=0.1)
```

# Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

# Stemmed
abs_s = CountVectorizer(max_df=0.95, min_df=0.05)
rel_s = TfidfVectorizer(max_df=0.95, min_df=0.05,use_idf=False, norm='l1')
tfidf_s = TfidfVectorizer(max_df=0.95, min_df=0.05, smooth_idf=False)
bayes_s = MultinomialNB()

#Lemmatized

abs_l = CountVectorizer(max_df=0.95, min_df=0.05)
rel_l = TfidfVectorizer(max_df=0.95, min_df=0.05,use_idf=False, norm='l1')
tfidf_l = TfidfVectorizer(max_df=0.95, min_df=0.05, smooth_idf=False)
bayes_l = MultinomialNB()
```

## Absolute Frequency

```python
# Stemmed 

X_train_stem_abs = abs_s.fit_transform(X_train_stem).toarray()
X_test_stem_abs = abs_s.transform(X_test_stem).toarray()

bayes_s.fit(X_train_stem_abs, y_train_stem)

y_pred_stem_abs = bayes_s.predict(X_test_stem_abs)

from sklearn.metrics import accuracy_score, classification_report
print('Train accuracy: ', bayes_s.score(X_train_stem_abs, y_train_stem))
print('Test accuracy: ', accuracy_score(y_pred_stem_abs, y_test_stem))
```

```python
# Lemmatized 

X_train_lemma_abs = abs_l.fit_transform(X_train_lemma).toarray()
X_test_lemma_abs = abs_l.transform(X_test_lemma).toarray()

bayes_l.fit(X_train_lemma_abs, y_train_lemma)

y_pred_lemma_abs = bayes_l.predict(X_test_lemma_abs)

print('Train accuracy: ', bayes_l.score(X_train_lemma_abs, y_train_lemma))
print('Test accuracy: ', accuracy_score(y_pred_lemma_abs, y_test_lemma))
```

## TF-IDF

```python
# Stemmed

X_train_stem_tfidf = tfidf_s.fit_transform(X_train_stem).toarray()
X_test_stem_tfidf = tfidf_s.transform(X_test_stem).toarray()

bayes_s.fit(X_train_stem_tfidf, y_train_stem)

y_pred_stem_tfidf = bayes_s.predict(X_test_stem_tfidf)

print('Train accuracy: ', bayes_s.score(X_train_stem_tfidf, y_train_stem))
print('Test accuracy: ', accuracy_score(y_pred_stem_tfidf, y_test_stem))
```

```python
# Lemmatized 

X_train_lemma_tfidf = tfidf_l.fit_transform(X_train_lemma).toarray()
X_test_lemma_tfidf = tfidf_l.transform(X_test_lemma).toarray()

bayes_l.fit(X_train_lemma_tfidf, y_train_lemma)

y_pred_lemma_tfidf = bayes_l.predict(X_test_lemma_tfidf)

print('Train accuracy: ', bayes_l.score(X_train_lemma_tfidf, y_train_lemma))
print('Test accuracy: ', accuracy_score(y_pred_lemma_tfidf, y_test_lemma))
```
